```python
# app/code/faiss/test.py

import requests
import json

# Базовые URL
auth_base_url = "http://127.0.0.1:5000/auth"
gpt_base_url = "http://127.0.0.1:5000/gpt"
headers = {"Content-Type": "application/json"}
```

```python
# Получение JWT

payload = {
    "api_token": "7a4e8f1d6c9b0a3e5f78f3d9a2c7e4b1a6f9d0e5c2bd2c4b8a1e9"
}
response = requests.post(f"{auth_base_url}/token", json=payload, headers=headers)
if response.status_code != 200:
    raise Exception(f"Ошибка авторизации: {response.status_code} {response.text}")
jwt_token = response.json()["access_token"]
print("JWT получен:", jwt_token)
```

```python
# Отправка запроса к GPT

gpt_payload = {
    "prompt": "Напиши короткий стих про весну",
    "temperature": 0.7,
    "max_tokens": 100
}
gpt_headers = {"Authorization": f"Bearer {jwt_token}", "Content-Type": "application/json"}
gpt_response = requests.post(gpt_base_url, json=gpt_payload, headers=gpt_headers)
print("POST /gpt — статус:", gpt_response.status_code)
try:
    answer = gpt_response.json()
    print("Ответ GPT:")
    print(json.dumps(answer, ensure_ascii=False, indent=2))
except json.JSONDecodeError:
    print("Ошибка декодирования JSON:", gpt_response.text)
```
