```python
# app/core/auth/test.ipynb

# Импорт зависимостей
import requests
import json

# Базовая конфигурация
base_url = "http://127.0.0.1:5000/auth"
headers = {"Content-Type": "application/json"}
```

```python
# Получение JWT токена

# Тело запроса
payload = {"api_token": "7a4e8f1d6c9b0a3e5f78f3d9a2c7e4b1a6f9d0e5c2bd2c4b8a1e9"}

# POST-запрос к эндпоинту /auth/token
response = requests.post(f"{base_url}/token", json=payload, headers=headers)

# Проверяем статус ответа
if response.status_code == 200:
    # Разбираем JSON
    data = response.json()
    print("Ответ backend:", json.dumps(data, indent=2))

    # Извлекаем токен и время жизни
    jwt_token = data["access_token"]
    expires_in = data.get("expires_in")

    print("JWT:", jwt_token)
    print("Время жизни токена (сек):", expires_in)
else:
    print(f"Ошибка {response.status_code}: {response.text}")
```
