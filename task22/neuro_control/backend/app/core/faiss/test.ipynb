```python
# app/code/faiss/test.py

import requests
import json

# Базовые URL
auth_base_url = "http://127.0.0.1:5000/auth"
faiss_base_url = "http://127.0.0.1:5000/faiss"
headers = {"Content-Type": "application/json"}
```

```python
# Получение JWT

payload = {
    "api_token": "7a4e8f1d6c9b0a3e5f78f3d9a2c7e4b1a6f9d0e5c2bd2c4b8a1e9"
}
response = requests.post(f"{auth_base_url}/token", json=payload, headers=headers)
if response.status_code != 200:
    raise Exception(f"Ошибка авторизации: {response.status_code} {response.text}")
jwt_token = response.json()["access_token"]
print("JWT получен:", jwt_token)
```

```python
# POST /faiss/build

build_url = f"{faiss_base_url}/build"
build_headers = {"Authorization": f"Bearer {jwt_token}", "Content-Type": "application/json"}
build_payload = {
    "text_file": "audio.txt",
    "chunk_size": 500
}
build_response = requests.post(build_url, json=build_payload, headers=build_headers)
print("POST /faiss/build — статус:", build_response.status_code)
print("Ответ:", json.dumps(build_response.json(), indent=2, ensure_ascii=False))
```

```python
# POST /faiss/search

search_url = f"{faiss_base_url}/search"
search_headers = {"Authorization": f"Bearer {jwt_token}", "Content-Type": "application/json"}
search_payload = {
    "text_file": "audio.txt",
    "query": "Как настроить CRM?",
    "k": 3
}
search_response = requests.post(search_url, json=search_payload, headers=search_headers)
print("POST /faiss/search — статус:", search_response.status_code)
print("Результаты поиска:", json.dumps(search_response.json(), indent=2, ensure_ascii=False))
```

```python
# POST /faiss/rebuild

rebuild_url = f"{faiss_base_url}/rebuild"
rebuild_headers = {"Authorization": f"Bearer {jwt_token}", "Content-Type": "application/json"}
rebuild_payload = {
    "text_file": "audio.txt",
    "chunk_size": 500
}
rebuild_response = requests.post(rebuild_url, json=rebuild_payload, headers=rebuild_headers)
print("POST /faiss/rebuild — статус:", rebuild_response.status_code)
print("Ответ:", json.dumps(rebuild_response.json(), indent=2, ensure_ascii=False))
```
